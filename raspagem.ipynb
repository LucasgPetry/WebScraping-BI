```python
import pandas as pd
import requests 
import numpy as np 
from bs4 import BeautifulSoup
```

```python
url = 'https://www.cnnbrasil.com.br/esportes/futebol/tabela-brasileirao-serie-a-2023/'
```

```python
tabela = pd.read_html(url)
display(tabela)
```

```python
new_tabela = np.reshape(tabela, (20,9))
print(new_tabela)
```

```python
df_tabela = pd.DataFrame(new_tabela)
df_tabela
```

```python
df_tabela = df_tabela.rename(
    columns = {
        0: "Classificação",
        1: "Pontuação", 
        2: "Jogos", 
        3: "Vitórias",
        4: "Empates", 
        5: "Derrotas", 
        6: "Gols Marcados", 
        7: "Gols cedidos", 
        8: "Saldo de Gols"
    }
)
```

```python
df_tabela
```

```python
print(df_tabela['Classificação'].unique())
```

```python
df_tabela['Classificação'] = df_tabela['Classificação'].str[slice(3,-5)]
```

```python
df_tabela
```

```python
print(df_tabela['Classificação'].unique())
```

```python
df_tabela.index = range(1, len(df_tabela) + 1)
```
